```python
import numpy as np
import qpmr
import qpmr.plot
from qpmr.core.quasipolynomial import create_normalized_delay_difference_eq
```

```python
coefs = np.array([
    [0, 1],
    [0, 1.4142],
    [0, 1.0000],
    [0, 2.2361],
    [1, 0],
    [0, 2.2361],
    [1, 0],
    [1, 0],
    [1, 0],
    [0, 0],
])
delays = np.array([0., np.sqrt(2), 1.6, np.sqrt(5), 8.0987, np.sqrt(10)*2+4] + [10+i for i in range(1,5)])
```

```python
roots, meta = qpmr.qpmr(coefs, delays)
```

```python
lb, ub = qpmr.bounds_neutral_strip(coefs, delays)
```

```python
x, y, mask = qpmr.distribution_diagram(coefs, delays)
qpmr.plot.spectrum_distribution_diagram(x, y, mask)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
qpmr.plot.roots(roots, ax=ax)
ax.axvline(lb, color="blue")
ax.axvline(ub, color="blue")
```
