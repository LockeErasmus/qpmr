```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import qpmr
import qpmr.plot
```

```python
region = [-0.9, 0.2, 0, 500]
coefs = np.array([[-1/4, 1.],
                  [1/3, -3/4],
                  [0, 1/2]])
```

```python
line_width = 252.0 / 72 # 1 pt = 1/72 inches
# plt.rcParams.update({
#     "text.usetex": False,
#     #"font.family": "Helvetica"
# })

filled_marker_style = dict(marker='o', markersize=20,
                           color='darkgrey',
                           markerfacecolor='none',
                           markerfacecoloralt='none',
                           markeredgecolor='darkgrey')
```

```python
fig, ax = plt.subplots(figsize=(line_width, 3))
delays = np.array([0, 1, 2.])
roots1, meta1 = qpmr.qpmr(region, coefs, delays)
sol1 = qpmr.safe_upper_bound_diff(coefs, delays)
qpmr.plot.roots(roots1, ax=ax)
ax.axvline(sol1.root, color="blue")
plt.tight_layout()
ax.set_ylim((0,500))
ax.set_xlim((-0.9, 0.2))
```

```python
delays = np.array([0, 1, 2.05])
roots2, meta2 = qpmr.qpmr(region, coefs, delays)
fig, ax = plt.subplots(figsize=(line_width, 3))
sol2 = qpmr.safe_upper_bound_diff(coefs, delays)
qpmr.plot.roots(roots2, ax=ax)
ax.axvline(sol2.root, color="blue")
plt.tight_layout()
ax.set_ylim((0,500))
ax.set_xlim((-0.9, 0.2))
```

```python
delays = np.array([0, 1, 2.005])
roots3, meta3 = qpmr.qpmr(region, coefs, delays)
fig, ax = plt.subplots(figsize=(line_width, 3))
sol3 = qpmr.safe_upper_bound_diff(coefs, delays)
qpmr.plot.roots(roots3, ax=ax)
ax.axvline(sol3.root, color="blue")
plt.tight_layout()
ax.set_ylim((0,500))
ax.set_xlim((-0.9, 0.2))
```

```python
fig, ax = plt.subplots(figsize=(line_width, 3))
tol = 1e-6

ax.add_patch(
    patches.Rectangle((0, -1), 10, 600, linewidth=1, edgecolor=None, facecolor='red', alpha=0.25)
)

for roots, meta, color in zip([roots1, roots2, roots3], [meta1, meta2, meta3], ["g", "k", "b"]):
    ax.scatter(
        np.real(roots),
        np.imag(roots),
        marker="+",
        color=color,
        linewidths=0.5,
    )

ax.set_ylim((0,500))
ax.set_xlim((-0.9, 0.2))
ax.set_xlabel(r"$\Re (\lambda)$")
ax.set_ylabel(r"$\Im (\lambda)$")
plt.tight_layout()
```
